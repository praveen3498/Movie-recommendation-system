```python
import pandas as pd
```

```python
import difflib
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('/content/drive/MyDrive/movies.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.fillna('',inplace=True)
```

```python
data.isnull().sum()
```

```python
imp_data=data['genres'] +" "+ data['keywords'] +" "+ data['tagline'] +" "+ data['title'] +" "+ data['cast']
```

```python
print(imp_data)
```

```python
vectorizer=TfidfVectorizer(stop_words='english',lowercase=True,min_df=1)
```

```python
final=vectorizer.fit_transform(imp_data)
```

```python
print(final)
```

```python
similarity=cosine_similarity(final)
```

```python
similarity.shape
```

```python
movie=input('enter a movie name : ')
```

```python
all_movies=data['title'].values.tolist()
```

```python
print(all_movies)
```

```python
close_match=difflib.get_close_matches(movie,all_movies)
```

```python
print(close_match)
```

```python
close_match=close_match[0]
```

```python
index_of_movie=data[data['title']==close_match]['index'].values[0]
```

```python
print(index_of_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_movie]))
```

```python
print(similarity_score)
```

```python
sorted_similarity=sorted(similarity_score, key=lambda x:x[1], reverse=True )
```

```python
print(sorted_similarity)
```

```python
for i in range(20):
  total=data[data['index']==sorted_similarity[i][0]]['title'].values[0]
  print('---------------------------------------')
  print(total)
```
